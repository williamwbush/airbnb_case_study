```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ab = pd.read_csv('AB_NYC_2019.csv')
ab.head()
```

```python
ab["booked"] = 365 - ab["availability_365"]
ab_booked = ab.groupby(["host_id", "host_name"]).sum()["booked"].sort_values(ascending = False)
ab_booked

# Blueground, Sonder (NYC), and Kazuya have the most currently booked nights with 25796, 20767, and 17358 booked nights respectively.
```

```python
# How many neighborhood groups are available and which shows up the most?

ab_ngcount = ab.groupby("neighbourhood_group", as_index=False).count().sort_values("id", ascending=False).reset_index(drop=True)
print(len(ab_ngcount))
print(ab_ngcount.loc[0][0])
ab_ngcount

# There are five neighborhood groups and Manhattan appears the most
```

```python
# Are private rooms the most popular in manhattan?

ab_man_rooms = ab.groupby(["neighbourhood_group","room_type"]).count().loc['Manhattan'].sort_values("id", ascending=False)
ab_man_rooms

# Entire home/apt is the most popular in Manhattan
```

```python
# Which hosts are the busiest and based on their reviews?

ab_host_busy = ab.groupby(["host_id","host_name"]).count().sort_values('number_of_reviews', ascending=False)["number_of_reviews"]
ab_host_busy

# Sonder (NYC), Blueground and Kara are the busiest host based on number of reviews.
```

```python
#Which neighorhood group has the highest average price?

ab_ng_hap = ab.groupby("neighbourhood_group").mean().sort_values("price", ascending=False)['price']
ab_ng_hap

# Manhattan has the highest average price
```

```python
# Which neighbor hood group has the highest total price?

# Finding the sum of all prices in each neighborhood group

ab_htp = ab.groupby("neighbourhood_group").sum().sort_values("price", ascending=False)["price"]
ab_htp

# Manhattan has the highest sum of listing prices


# Finding the neighborhood with the most expensive listing

ab_hp = ab.sort_values("price", ascending=False)[["neighbourhood_group", "price"]].reset_index(drop=True)
ab_hp

# The most expensive listing is a tie between Queens, Brooklyn, and Manhattan at $10,000.
```

```python
#Which top 5 hosts have the highest total price?

ab_host_htp = ab.groupby(["host_id","host_name"]).sum().sort_values("price", ascending=False).head(5)["price"]
ab_host_htp

# Sonder (NYC), Blueground, Sally, Red Awning, and Kara have the highest sum of listing prices
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

ab_host_reviews = ab.groupby(["host_id","host_name"], as_index = False).count().sort_values("number_of_reviews", ascending=False)
ab_host_reviews_100 = ab_host_reviews[ab_host_reviews["number_of_reviews"] >= 100][["host_id", "host_name", "number_of_reviews"]]
ab_host_reviews_100
```

```python
ab_host_avail = ab.groupby(["host_id","host_name"], as_index=False).sum()[["host_id", "host_name", "availability_365"]]
ab_host_avail
```

```python
ab_host_100_avail = ab_host_avail.merge(ab_host_reviews_100, on = "host_id", how = "inner")
ab_host_100_avail

# All hosts with a review count of 100 or more have listings available in the coming year
```

```python
# What host has the highest total of prices and where are they located?

ab_host_htp.head(1)
```

```python
ab.set_index("host_id").loc[219517861].groupby("neighbourhood_group").count()

# Sonder (NYC) has the highest total of prices ($82,795) and is located in Manhattan.
```

```python
# When did Danielle from Queens last receive a review?

ab_danielle = ab[(ab["host_name"] == "Danielle") & (ab["neighbourhood_group"] == "Queens")].sort_values("last_review", ascending=False)
ab_danielle[["name", "host_name", "neighbourhood_group", "last_review"]]

# Danielle last receieved a review on July 8th, 2019.
```
